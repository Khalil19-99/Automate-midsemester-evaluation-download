# Task

 

1. Implement control, but state observer (x can't be used by the controller, but $y = Cx$ can be)
2. Simulate the system with the observer and the controller
3. Simulate from different initial positions



```python
from control import lqr

def my_generateLQRTable(A_table, B_table, Q_table, R_table):
    count = A_table.shape[0]
    n = A_table.shape[2]
    m = B_table.shape[2]

    K_table = np.zeros((count,m,n))

    for i in range(count):
        K, S, CLP =  lqr(A_table[i], B_table[i], Q_table[i], R_table[i])
        K_table[i] = K
        
    return K_table
```

```python
def generate_LQR_LTable(A_table, C, Q_table, R_table):
    count = A_table.shape[0]
    n = A_table.shape[1]
    m = C.shape[0]

    L_table = np.zeros((count, n, m))
    for i in range(count):
        LT, _, _ = lqr(A_table[i].T, C.T, Q_table[i], R_table[i])
        L_table[i] = LT.T
    return L_table
```

```python
def observer_error_dynamics_closed_loop(A_table, C, L_table):
    count = A_table.shape[0]
    m = A_table.shape[1]
    n = A_table.shape[2]

    AA_table = np.zeros((count, m, n))
    for i in range(count):
        AA = A_table[i] - L_table[i] @ C
        AA_table[i] = AA
    return AA_table
```

```python
C = np.concatenate((np.eye(n), np.zeros((n, n))), axis=1) #y = C*x

Q = 1e4*np.eye(2 * n)
R = np.eye(handlerGeneralizedCoordinatesModel.dofControl)
count = A_table.shape[0]

#Controller gains
K_table = my_generateLQRTable(A_table, B_table, np.tile(Q, [count,1, 1]), np.tile(R, [ count, 1, 1]))
#Closed-loop system
AA_table, cc_table = generateCloseLoopTable(A_table, B_table, c_table, K_table, x_table, u_table)
```

```python
# Computing observers table
L_table = generate_LQR_LTable(A_table, C, np.tile(Q, [count,1, 1]), np.tile(R, [ count, 1, 1]))

# Computing closed-loop observer error dynamics
Eps_dynamics = observer_error_dynamics_closed_loop(A_table, C, L_table)
# print(Eps_dynamics)

# Err dynamics simulation
#init_noise = (np.random.rand(initialPosition.shape[0]) - 0.5) / 2 + 1
init_noise = np.ones(initialPosition.shape[0])

x0_obs = np.hstack((initialPosition * init_noise, np.zeros(initialPosition.shape[0])))
err_solution_tape = np.empty((14,1))
count = x_table.shape[0]
ode_fnc_err_handle = ClosedLoopLinearSystemOdeFunctionHandler(Eps_dynamics, np.zeros(cc_table.shape), timeTable)

for i in range(1,count):
    x_desired = x_table[i]
    t_init = timeTable[i-1]
    t_final = timeTable[i]
    eps0 = x0_obs - x_desired

    err_sol = solve_ivp(ode_fnc_err_handle, [t_init, t_final], eps0, t_eval=timeTable[i-1:i],method="RK45")

    err_solution_tape = np.hstack((err_solution_tape, err_sol.y))
    x0_obs = x_desired + err_solution_tape[:, -1]

# plot graphs

ax = plotGeneric(timeTable,err_solution_tape.T,figureTitle="Error dynamics",ylabel="ODE")
```

```python
for Eps in Eps_dynamics:
    eigenvalues, _ = np.linalg.eig(Eps)
    if np.any(eigenvalues.real > 0):
        print(f"Ooops... {eigenvalues}")
        break 
```

```python
#Initial positions
x0 = np.hstack((initialPosition, np.zeros(initialPosition.shape[0])))

#Simulation
ode_fnc_handle = ClosedLoopLinearSystemOdeFunctionHandler(AA_table, cc_table, timeTable)
sol = solve_ivp(ode_fnc_handle, [0, tf], x0, t_eval=timeTable,method="RK45")

# plot graphs
time_table_0 = sol.t
solution_tape = sol.y.T

ax = plotGeneric(time_table_0,solution_tape,figureTitle="",ylabel="ODE")
# ax = plotGeneric(timeTable,x_table,ylabel="linearmodel",old_ax = ax, plot=True)

ax = plotGeneric(timeTable,solution_tape[:,:n],figureTitle="position",ylabel="q", plot=True)
ax = plotGeneric(timeTable,solution_tape[:,n:2*n],figureTitle="velocity",ylabel="v", plot=True)

with open('anim_array.npy', 'wb') as f:
    np.save(f, solution_tape[:,:n])
```
